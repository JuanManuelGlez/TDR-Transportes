#**Determinar categorías y rangos típicos**

En este caso, para realizar este paso, vamos a estudiar los datos de las casetas y de las operaciones de liverpool. Lo anterior, para identificar rangos que pueden ser intepretados como categoría y rangos.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

columnas_numericas = ['Line Haul Charge', 'Accessorial Charge', 'Order Total Charge']
for columna in columnas_numericas:
    plt.figure(figsize=(10, 6))
    sns.histplot(DF_LV[columna], kde=True)
    plt.title(f'Distribución de {columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.show()

plt.figure(figsize=(12, 8))
sns.countplot(data=DF_LV, y='Order Status')
plt.title('Frecuencia de Estado de Orden')
plt.xlabel('Frecuencia')
plt.ylabel('Estado de Orden')
plt.show()
```

```python
correlacion = DF_LV.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor de Correlación')

plt.show()
```

# Distribución y correlación de datos
En este caso, vemos que los datos e las variables Order Total Charge y Line Houl Charge, están cargados hacía la izquierda, asimismo, existe una correlación de datos entre las variables que muestras un cargo y aquellas que muestran la cantidad recorrida.
