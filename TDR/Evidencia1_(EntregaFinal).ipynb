#**4. Utilizar one-hot encoding para las variables categóricas que consideren necesarias.**

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np


encoder = OneHotEncoder(sparse=False)

# Seleccionar las columnas a codificar
columns_to_encode = ['Municipo origen', 'Municipo destino']

# Ajustar y transformar los datos seleccionados con el encoder
encoded_data = encoder.fit_transform(df[columns_to_encode])
columns_encoded = encoder.get_feature_names_out(columns_to_encode)

# Convertir los datos codificados en un DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=columns_encoded)

# Resetear el índice en 'df' para evitar problemas al concatenar
df.reset_index(drop=True, inplace=True)

# Concatenar el DataFrame original con las nuevas columnas codificadas
df_expanded = pd.concat([df, encoded_df], axis=1)
print(df_expanded.head())
```

**Al utilizar one-hot encoding para los municipios de origen y destino, establecemos una base sólida que mejora nuestro proceso de análisis relacionado con los telepeajes en cada viaje. Esta técnica nos permite desglosar y organizar los datos de manera que podemos identificar con precisión dónde y cuándo ocurren los cruces de telepeaje, facilitando la comparación entre los telepeajes comprados y los realmente utilizados. Al convertir las ubicaciones en columnas binarias, obtenemos una visión clara de los patrones de viaje, lo que nos ayuda a asegurar que el cobro a los clientes por el uso de casetas sea completo y refleje los costos incurridos.
```python
#Detectar valores atípicos
import numpy as np
import matplotlib.pyplot as plt

def detect_outliers_IQR(df, column_name):
    Q1 = np.percentile(df[column_name], 25)
    Q3 = np.percentile(df[column_name], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[~((df[column_name] <= upper_bound) & (df[column_name] >= lower_bound))]
    return outliers

outliers_km = detect_outliers_IQR(df, 'km totales')
print(outliers_km)
outliers_caseta_IQR = detect_outliers_IQR(casetasL, 'Costo Caseta')
print(outliers_caseta_IQR)

# Graficos
plt.figure(figsize=(10, 6))
plt.boxplot(df['km totales'], vert=False)
plt.xlabel('km totales')
plt.title('Valores atípicos en km totales')
plt.show()

plt.figure(figsize=(10, 6))
plt.boxplot(casetasL['Costo Caseta'], vert=False)
plt.xlabel('Costo Caseta')
plt.title('Valores atípicos en Costo Caseta')
plt.show()
```
