```python
import pandas as pd 
```

```python
DF_casetas=pd.read_excel('BD_TR.xlsx',sheet_name= 'CasetasL')
DF_casetas.head()
```

```python
columnas_no_relevantes = ['EC', 'Mes']

# Eliminar columnas no relevantes
DF_casetas = DF_casetas.drop(columnas_no_relevantes, axis=1)
DF_casetas.head()
```

```python
dispach = pd.read_excel('Liverpool_final.xlsx')
dispach.head()
```

```python
#vamos a hacer una lista con las columnas relevantes para eliminar todas las demas 

columnas = ['ord_number', 'billto_cmp_id','ord_startdate', 'ord_completiondate','ord_revtype4', 'origin_cty_nmstct','dest_cyt_nmstct','ord_tractor','ord_totalmiles']

dispach = dispach[columnas]

#Cambniamos el nombre de las columnas 
dispach.columns = ['Numero de orden', 'Cliente que paga', 'Fecha de inicio', 'Fecha de finalizacion', 'Equipo colaborativo', 'Municipo origen', 'Municipo destino', '# de Tractor', 'Millas totales']

dispach.head()
```

```python

dispach['Fecha de inicio'] = dispach['Fecha de inicio'].astype(str)
dispach['Fecha de finalizacion'] = dispach['Fecha de finalizacion'].astype(str)

dispach[['Fecha_inicio', 'Fecha_inicio_hora']] = dispach['Fecha de inicio'].str.split(' ', expand=True)
dispach[['Fecha_fin', 'Fecha_fin_hora']] = dispach['Fecha de finalizacion'].str.split(' ', expand=True)

```

```python
dispach=dispach.drop(['Fecha de inicio','Fecha de finalizacion','Fecha_inicio','Fecha_inicio_hora','Fecha_fin_hora'], axis=1)
```

```python
dispach.rename(columns={'# de Tractor': 'Unidad'}, inplace=True)
dispach.rename(columns={'Fecha_fin': 'Fecha'}, inplace=True)
```

```python
dispach['Fecha'] = pd.to_datetime(dispach['Fecha'], errors='coerce')
DF_casetas['Fecha'] = pd.to_datetime(DF_casetas['Fecha'], errors='coerce')
```

```python
DF_unido= pd.merge(dispach,DF_casetas, on=['Unidad','Fecha'], how='inner') 
```
