```python
import pandas as pd
```

```python
#leemos el dataframe solo con las columnas que nos interesan
df = pd.read_excel('Solo_liverpool2.xlsx', usecols=['Número de orden', 'Nombre de la empresa', 'Cliente que paga', 'Fecha de inicio del viaje', 'Fecha fin del viaje', 'Municipo origen', 'Municipo destino', '# de Tractor', 'km totales'])

df.head()
```

```python
#eliminamos la primera fila 

df = df.drop(df.index[0])

df.head()
```

```python
casetasL = pd.read_excel('BD_TR.xlsx', sheet_name='CasetasL')
BDL = pd.read_excel('BD_TR.xlsx', sheet_name='BDL')

casetasL.head()
BDL.head()

#Quitamos la columna BDL Unnamed: 2
BDL = BDL.drop(columns=['Unnamed: 2'])
BDL.head()
```

```python
#Renombramos las columnas de BDL a TAG
BDL = BDL.rename(columns={'Tag': 'TAG'})

#Hacemos un Merge de BDL con casetasL
casetasL = pd.merge(casetasL, BDL, on='TAG', how='left')

casetasL.head()
```

```python
# hacemos un grupby de df por nombre de la empresa con la suma de los km totales

LiverpoolGruped = df.groupby('Nombre de la empresa')['km totales'].sum().reset_index()

LiverpoolGruped.head()
```
